```python
import cns
import cns.data_utils as cdu
```

```python
samples_df, cns_df = cdu.main_load("COSMIC")
```

```python
cns_df = cns.add_total_cn(cns_df)
lusc_df = cns.select_cns_by_type(cns_df, samples_df, "LUSC")
lusc_one_df = cns.cns_head(lusc_df, 1)
rank_list = lusc_one_df.sort_values(by="total_cn", ascending=False)["name"].to_list()
top_6_types = samples_df["type"].value_counts().index[:6].to_list()
```

```python
question = "Below is a list of COSMIC {gene_names} ordered by descending total copy number in a cell. Your task is to identify the cancer type from the response list.\n\n"
question += "{gene_names}: " + " ".join(rank_list) + "\n\n"
question += "{response}: " + " ".join(top_6_types) + "\n\n"
question += "Answer:"
```

```python
print(question)
```
