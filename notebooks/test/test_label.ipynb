```python
from cns.process.binning import bin_by_breaks, mean_bins, add_cns_loc, get_breakpoints

pcawg_cns_df, _ = load_cns_out("PCAWG_cns_imp.tsv")
sample_ids = pcawg_cns_df.index.unique()
pcawg_cns_10_df = pcawg_cns_df[pcawg_cns_df.index.isin(sample_ids[:10])].copy()
pcawg_10_bin_df = bin_by_breaks(pcawg_cns_10_df, get_breakpoints(str(int(1e5))))
pcawg_10_groups_df = mean_bins(add_cns_loc(pcawg_10_bin_df))
max_cn = 10
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 2), dpi=150)
line_plot(ax, pcawg_10_groups_df, "major_cn", "blue", "MAJOR PCAWG BLUE", .5, 2)
ax.legend()
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 2), dpi=150)
scatter_plot(ax, pcawg_10_groups_df, "major_cn", "blue", "MAJOR PCAWG BLUE", .5, 2)
ax.legend()
```
