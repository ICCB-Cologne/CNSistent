```python
from cns.process.binning import *
from cns.process.breakpoints import *
from cns.process.segments import *
from cns.utils.assemblies import hg19
from cns.analyze.genome import *
from cns.analyze.heatmap import *

# auto reload modules
%load_ext autoreload
%autoreload 2
```

```python
cna_df = pd.read_csv("./out/PCAWG_cna_imp.tsv", sep="\t")
cna_df = cna_df[cna_df["sample_id"].isin(cna_df["sample_id"].unique()[:20])]
cna_df.head()
```

```python
genes = pd.read_csv("./data/COSMIC_consensus_genes.tsv", sep="\t")
genes.head()
```

```python
segments = regions_to_segments(genes)
segments[:5]
```

```python
cosmic_bins = bin_by_segments(cna_df, segments)
cosmic_bins = add_derived(cosmic_bins, hg19)
cosmic_bins.head()
```

```python
grouped = mean_bins(cosmic_bins, hg19)
grouped.head()
```

```python
fig_CN_heatmap(cosmic_bins, "CNs COSMIC", dpi=300);
```

```python
# join grouped and genes on chrom, start, end
grouped = grouped.reset_index()
merged = pd.merge(grouped, genes, on=["chrom", "start", "end"])
merged.head()
```

```python
selection = merged.query("total_cn > 6")
selection
```

```python

max_cn = merged["total_cn"].max() + .5

fig, ax = fig_manhattan(merged, hg19, "COSMIC", max_cn=max_cn)

# add text labels for selection["gene"]
for i, row in selection.iterrows():
    ax.text(row["cum_mid"], row["total_cn"], row["gene"], fontsize=8)
```
